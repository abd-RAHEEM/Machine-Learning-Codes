```python
import pandas as pd 
import numpy as np
```

```python
org_data=pd.read_csv("datasets/Qinspection.csv")
org_data.head(5)
```

```python
features=org_data.iloc[:,0:6]
features
```

```python
classLabel=org_data.iloc[:,-1]
classLabel.head(5)
```

```python
from sklearn.preprocessing import StandardScaler 
stds=StandardScaler()
tfeatures=stds.fit_transform(features)
```

```python
tdf=pd.DataFrame(tfeatures)
features=['feature1', 'feature2', 'feature3', 'feature4','feature5','feature6']
tdf.columns=features
tdf.head(5)
```

```python
cor_matrix=tdf.corr().abs()
print(cor_matrix)
```

```python
upper_tri=cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
print(upper_tri)
```

```python
to_drop=[c for c in upper_tri.columns if any(upper_tri[c]>0.95) ]
print(to_drop)
```

```python
df1=tdf.drop(columns=to_drop,axis=1)
```

```python
dflabel=pd.DataFrame(classLabel)
new_data=pd.concat([df1,dflabel],axis=1)
print(new_data.head(5))
```
